# Spectral Data

APOGEE spectral data of 14370 stars with effective temperature from 2500K to 4000K. The spectral data can be read as following:

```python
import pickle
import matplotlib.pyplot as plt
```

```python
data_path = '/stow/weilingfeng/data/apogee/real_spectra.pkl'
with open(data_path, 'rb') as file:
    spectra = pickle.load(file)
```

The wavelength is a 1D numpy array of length 8575 and can be accessed by `spectra['wave']`. The flux is a 2D masked array of shape (14370, 8575) and can be accessed by `spectra['flux']`. The spectra of the i-th star can be plotted as following:

```python
i = 3   # plot the first spectrum
fig, ax = plt.subplots(figsize=(12, 4.5))
ax.plot(spectra['wave'], spectra['flux'][i], lw=0.7, alpha=0.8)
ax.set_xlabel('Wavelength (Å)')
ax.set_ylabel('Flux')
ax.set_title(f"{spectra['ID'][i]}")
plt.show()
```

Below is a detailed description of each column

| Column    | Description           | Dtype | Dimension     |
| ------    | -----------           | ----- | ---------     |
| `ID`      | APOGEE ID             | str   | (14334, )     |
| `Loc`     | APOGEE Location ID    | int   | (14334, )     |
| `RAJ2000` | Right Ascension       | float | (14334, )     |
| `DEJ2000` | Declination in        | float | (14334, )     |
| `SNR`     | Signal-to-noise ratio | float | (14334, )     |
| `HRV`     | Radial velocity       | float | (14334, )     |
| `e_HRV`   | Uncertainty of `HRV`  | float | (14334, )     |
| `Teff`    | Effective temperature | float | (14334, )     |
| `e_Teff`  | Uncertainty of `Teff` | float | (14334, )     |
| `logg`    | Surface gravity       | float | (14334, )     |
| `e_logg`  | Uncertainty of `logg` | float | (14334, )     |
| `Vsini`   | Rotational velocity   | float | (14334, )     |
| `[M/H]`   | Metallicity           | float | (14334, )     |
| `e_[M/H]` | Uncertainty of `[M/H]`| float | (14334, )     |
| `wave`    | Wavelength of spectra | float | (7514, )      |
| `flux`    | Flux of spectra       | float | (14334, 7514) |
| `lsf`     | Line Spread Function  | float | (25724, 43)   |
| `xlsf`    | Line Spread Function  | float | (43,)         |
